Corremos la simulación en el notebook

```python
# MAP
map_cfg = "../configs/1er_piso_ala_sur_alig.yaml"

# ROBOT
robot_cfg = "../configs/roomba.yaml"

# SENSOR
sensor_cfg = "../configs/lidar.yaml"
scans_scale_factor = 3

# TASKS
tasks = ["explore"]
tasks_cfgs = [f"../configs/{t}" for t in tasks]

# SIMULATION PARAMS
use_roomba = False
seed = 1234
max_duration = 3 * 60 # Duración máxima de la simulación en segundos
sample_time = 0.25 # Tiempo de muestreo de la simulación
xlims = (5, 15)
ylims = (12, 22)

s = Simulator(
    map_cfg=map_cfg,
    robot_cfg=robot_cfg,
    sensor_cfg=sensor_cfg,
    tasks_cfgs=tasks_cfgs,
    scans_scale_factor=scans_scale_factor,
    use_roomba=use_roomba,
    seed=seed,
    xlims=xlims,
    ylims=ylims,
    max_duration=max_duration,
    sample_time=sample_time,
).run()
```
